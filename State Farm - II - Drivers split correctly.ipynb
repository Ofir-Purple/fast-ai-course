## We'll start with the simplest model possible

```python
train_batches = get_batches(train_path, batch_size=batch_size)
valid_batches = get_batches(valid_path, batch_size=batch_size, shuffle=True)
```

```python
def get_plain_model(lr=1e-5, with_regularization=False):
    
    plain_model = Sequential()
    
    plain_model.add(BatchNormalization(axis=1, input_shape=train_batches.image_shape))
    plain_model.add(Flatten())
    if not with_regularization:
        plain_model.add(Dense(len(classes), activation='softmax'))
    else:
        plain_model.add(Dense(len(classes), activation='softmax', W_regularizer=l2(l=0.01)))

    opt = Adam(lr=lr)
    plain_model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])  
    
    return plain_model
```

```python
plain_model = get_plain_model(lr=1e-5)
```

```python
# As can be seen from the results below, we're majorly overfitting (which is to be expected). 
# We reached an accuracy of 0.9x and the validation accuracy flattened at around 0.26-0.27 after 4 epochs.

plain_model.fit_generator(
    train_batches,
    samples_per_epoch=train_batches.nb_sample,
    nb_epoch=10,
    validation_data=valid_batches,
    nb_val_samples=valid_batches.nb_sample,
)
```

```python
# Let's see if a conv layer can help here
def get_simple_conv_model(lr=1e-5):
    
    conv_model = Sequential()
    
    conv_model.add(BatchNormalization(axis=1, input_shape=train_batches.image_shape))
    
    conv_model.add(ZeroPadding2D())
    conv_model.add(Convolution2D(nb_filter=32, nb_row=3, nb_col=3, activation='relu'))
    conv_model.add(BatchNormalization(axis=1,))
    conv_model.add(MaxPooling2D())
    
    conv_model.add(ZeroPadding2D())
    conv_model.add(Convolution2D(nb_filter=64, nb_row=3, nb_col=3, activation='relu'))
    conv_model.add(BatchNormalization(axis=1,))
    conv_model.add(MaxPooling2D())

    conv_model.add(Flatten())
    conv_model.add(Dense(200, activation='relu'))
    conv_model.add(Dropout(0))
    conv_model.add(BatchNormalization())
    conv_model.add(Dense(len(classes), activation='softmax'))
    
    opt = Adam(lr=lr)
    conv_model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
    
    return conv_model
```

```python
conv_model = get_simple_conv_model()
```

```python
# From the below the valid_accuracy is smaller than with the plain model. I assume this is because there are too many
# parameters and the model is just overfitting. Cause, you know.. those are images.. 
# conv nets are supposed to help here.. 

conv_model.fit_generator(
    train_batches,
    samples_per_epoch=train_batches.nb_sample,
    nb_epoch=2,
    validation_data=valid_batches,
    nb_val_samples=valid_batches.nb_sample,
)
```

```python
conv_model.optimizer.lr = 0.01
conv_model.fit_generator(train_batches, samples_per_epoch=train_batches.nb_sample, nb_epoch=4,
    validation_data=valid_batches, nb_val_samples=valid_batches.nb_sample)
```
